## Часть I: О теореме Байеса

### Описание задачи "убийца":
1. Произошло убийство. На месте убийства найдена кровь, которая явно (будем считать, что с вероятностью 1) принадлежит настоящему убийце. Кровь принадлежит редкой группе, которая присутствует всего лишь у 1% населения. И так получилось, что у подсудимого как раз эта редкая группа крови!
Первым выходит прокурор и говорит: “Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен - всего 1%; значит, с вероятностью 99% он виновен, предлагаю осудить”. В чём не прав прокурор? Укажите, какие вероятности он оценил, а какие должен был оценить.
2. Затем выходит адвокат, объясняет ошибки прокурора и берёт слово сам: “В городе живёт миллион человек. Это значит, что примерно у 10000 из них такая группа крови. Значит, всё, что говорит нам эта кровь -- это то, что подсудимый совершил убийство с вероятностью 0.01%. Получается, что это свидетельство не просто не является строгим доказательством, а и вовсе даёт нам ничтожно малую вероятность того, что мой подзащитный виновен, так что предлагаю исключить его из материалов дела”. В чём не прав адвокат? Какие вероятности он оценил, а какие должен был оценить?

### Решение задачи "убийца":
Для начала проанализируем все факты, которые имеются, вследствии чего сделаем вывод, какие вероятности действительно нужно оценить, и оценим их. Во-первых введем два события:<br>
$A$ - подсудимый виновен (этот факт абсолютно эквивалентен тому, что именно его кровь нашли на месте убийства)<br>
$B$ - у подсудимого редкая группа крови (хотя этот факт и известен, будем пологать, что мы оцениваем вероятность того, что у какого-то любого человека кровь редкая)<br>

Теперь оценим априорную вероятность. Для этого сделаем допущение, что как убийца, так и подсудимый живут в городе, про который говорил адвокат и численность населения в котором $N = 10^6$. Итак:
$$
P(A) = \frac{1}{N} = \frac{1}{10^6}
$$

Далее моделируем нашу ситуацию с помощью формулы Байеса, учитывая тот факт, что найденная кровь принадлежит редкой группе (т е у убийцы 100% редкая группа крови):<br>
$$
P(A|B) = \frac{P(B|A) P(A)}{P(B)} = \frac{1 \cdot P(A)}{0.01} = 100 \cdot P(A)
$$

Таким образом можно видеть, что апостериорная вероятность в 100 раз больше априорной. А при подстоновке:
$$
P(A) = \frac{1}{10^6}
$$
получаем в итоге:
$$
P(A|B) = \frac{1}{10^4}
$$

Таким образом, факт "редкой крови" повышает шанс того, что подсудимый - убийца в сто раз, однако этот шанс все еще остается очень небольшим. Теперь разберемся с прокурором и адвокатом, которые пытаются отработать свои деньги не очень честным путем.
1. Утверждение "Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен - всего 1%" абсолютно верное, однако отсюда, мы можем сделать лишь такой вывод:
$$
P(B|\bar A) + P(\bar B|\bar A) = 1 => P(\bar B|\bar A) = 1 - P(B|\bar A) = 1 - 0.01 = 0.99
$$
т е вероятность того, что у подсудимого, если он невиновен, не редкая группа крови - 99%. Из этих фактов никаких выводов о виновности подсудимого мы сделать не можем, т к учитываем условие невиновности уже как данное.
2. В итоговой оценке он действительно прав. Однако он не упомянул, что без учета факта "редкой крови" вероятность (априорная) виновности подсудимого была в 100 раз меньше. Поэтому этот факт ни в коем случае нельзя исключать.

### Описание задачи "тест на коронавирус":

Врачи решили, что терпеть это никак нельзя; они готовы провести срочное исследование и улучшить качество теста. Но сил и денег хватит только для того, чтобы сократить один тип ошибок, т.е. уменьшить или только число false positives (когда тест выдаёт положительный результат у здорового человека), или только число false negatives (когда тест выдаёт отрицательный результат у больного).
Помогите медикам: постройте зависимости апостериорной вероятности болезни после положительного и отрицательного теста от обоих видов ошибок (например, в виде графиков), сделайте выводы и дайте рекомендации врачам -- на чём лучше сосредоточиться?

### Решение задачи "убийца":
Для начала введем такие события:<br>
$A$ - тест положительный<br>
$B$ - пациент болен<br>

Тогда выразим изменяющиеся аргументы:
$$
P(A|\bar B) = FP\text{ (false negative)}
$$
$$
P(\bar A|B) = FN\text{ (false positive)}
$$

Заранее произведем некоторые математические выкладки:
$$
P(A|B) + P(\bar A|B) = 1 => P(A|B) = 1 - P(\bar A|B) = 1 - FN
$$
$$
P(A|\bar B) + P(\bar A|\bar B) = 1 => P(\bar A|\bar B) = 1 - P(A|\bar B) = 1 - FP
$$

И найдем зависимость исследумых функций от аргументов:
$$
P1 = P(B|A) = \frac{P(A|B) \cdot P(B)}{P(A)}
= \frac{(1 - FN) \cdot P(B)}{P(A|B) \cdot P(B) + P(A|\bar B) \cdot P(\bar B)}
= \frac{(1 - FN) \cdot 0.01}{(1 - FN) \cdot 0.01 + FP \cdot 0.99} = \frac{1 - FN}{1 - FN + 99FP}
$$
$$
P2 = P(B|\bar A) = \frac{P(\bar A|B) \cdot P(B)}{P(\bar A)}
= \frac{FN \cdot P(B)}{P(\bar A|B) \cdot P(B) + P(\bar A|\bar B) \cdot P(\bar B)} 
= \frac{FN \cdot 0.01}{FN \cdot 0.01 + (1 - FP) \cdot 0.99} = \frac{FN}{99 + FN - 99FP}
$$<br>
Теперь посторим графики. Заранее оговоримся, что нам нужно максимизировать $P1$ или минимизировать $P2$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def p1(fn, fp):
    return (1 - fn) / (1 - fn + 99 * fp)

x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)

X, Y = np.meshgrid(x, y)
Z = p1(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
```

```python
def p2(fn, fp):
    return fn / (99 + fn - 99 * fp)

x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)

X, Y = np.meshgrid(x, y)
Z = p2(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
```

Для лучшего понимания стратегии уменьшения ошибок, построим оба графика в разрезе при фиксировании одного из значений fp или fn (берем текущее значение, т е 0.05).

```python
p1fp = lambda fp: (1 - 0.05) / (1 - 0.05 + 99 * fp)
p1fn = lambda fn: (1 - fn) / (1 - fn + 99 * 0.05)

x = np.linspace(0, 0.05, 100)

plt.plot(x, p1fp(x))
plt.plot(x, p1fn(x))
plt.scatter(0.05, p1fp(0.05), color="black")
plt.show()
```

```python
p2fp = lambda fp: 0.05 / (99 + 0.05 - 99 * fp)
p2fn = lambda fn: fn / (99 + fn - 99 * 0.05)

x = np.linspace(0, 0.05, 100)

plt.plot(x, p2fp(x))
plt.plot(x, p2fn(x))
plt.scatter(0.05, p2fp(0.05), color="black")
plt.show()
```

Из первого графика видно, что лучше уменьшать FP, т к после (т. е. левее) наших точкек (fp, fn = 0.05) для синего графика и (fp = 0.05, fn) для оранжевого графика, P1 будет возрастать значительно сильнее при уменьшении FP, нежели при уменьшении FN. Из второго - наоборот: видно что уменьшение FN будет лучше уменьшать P2. Однако предпочтение все же стоит отдать уменьшению FP, т к увеличение P1 гораздо значительнее, нежели уменьшение P2.
